```python
import requests

# 設置 API 密鑰
api_key = 'API_KEY'

# 設置搜索關鍵字和時間範圍
search_query = ''  # 搜索關鍵字
from_date = '2024-02-18'  # 起始日期
to_date = '2024-03-20'  # 結束日期

# 發送 HTTP 請求到 News API 的端點
url = f'https://newsapi.org/v2/everything?q={search_query}&from={from_date}&to={to_date}&apiKey={api_key}'
response = requests.get(url)  # 發送 HTTP GET 請求

# 檢查響應狀態碼
if response.status_code == 200:
    # 解析 JSON 響應
    data = response.json()  # 將響應轉換為 JSON 格式
    # 檢查是否存在文章數據
    if 'articles' in data:
        # 處理數據
        articles = data['articles']  # 提取文章數據
        for article in articles:
            title = article['title']  # 提取文章標題
            description = article['description']  # 提取文章描述
            url = article['url']  # 提取文章来源网址
            print(f'標題: {title}')
            print(f'描述: {description}')
            print(f'来源网址: {url}')  # 输出文章来源网址
            print('---')
    else:
        print('未找到文章。')  # 輸出未找到文章的消息
else:
    print('錯誤:', response.status_code)  # 輸出錯誤消息，包括狀態碼
```

```python
import tkinter as tk
from tkinter import scrolledtext
import requests

def fetch_news():
    # 获取用户输入的搜索关键字和时间范围
    search_query = entry_query.get()
    from_date = entry_from_date.get()
    to_date = entry_to_date.get()

    # 发送 HTTP 请求到 News API 的端点
    url = f'https://newsapi.org/v2/everything?q={search_query}&from={from_date}&to={to_date}&apiKey={api_key}'
    response = requests.get(url)

    # 清空之前的输出
    text_output.delete(1.0, tk.END)

    # 检查响应状态码
    if response.status_code == 200:
        # 解析 JSON 响应
        data = response.json()
        # 检查是否存在文章数据
        if 'articles' in data:
            # 处理数据
            articles = data['articles']
            for article in articles:
                title = article['title']
                description = article['description']
                url = article['url']
                text_output.insert(tk.END, f'標題: {title}\n描述: {description}\n來源網址: {url}\n\n')
        else:
            text_output.insert(tk.END, '未找到文章。\n')
    else:
        text_output.insert(tk.END, f'錯誤: {response.status_code}\n')

# 创建主窗口
root = tk.Tk()
root.title("新聞搜索應用")

# 创建用户输入框和标签
label_query = tk.Label(root, text="搜索關鍵字:")
label_query.grid(row=0, column=0, padx=5, pady=5)
entry_query = tk.Entry(root)
entry_query.grid(row=0, column=1, padx=5, pady=5)

label_from_date = tk.Label(root, text="起始日期:")
label_from_date.grid(row=1, column=0, padx=5, pady=5)
entry_from_date = tk.Entry(root)
entry_from_date.grid(row=1, column=1, padx=5, pady=5)

label_to_date = tk.Label(root, text="結束日期:")
label_to_date.grid(row=2, column=0, padx=5, pady=5)
entry_to_date = tk.Entry(root)
entry_to_date.grid(row=2, column=1, padx=5, pady=5)

# 创建按钮
button_search = tk.Button(root, text="搜索新聞", command=fetch_news)
button_search.grid(row=3, column=0, columnspan=2, padx=5, pady=5)

# 创建文本框来显示新闻
text_output = scrolledtext.ScrolledText(root, width=100, height=100)
text_output.grid(row=4, column=0, columnspan=2, padx=5, pady=5)

# 运行主循环
root.mainloop()

```
