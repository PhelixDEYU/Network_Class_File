```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('adult.csv')

# Display the first few rows of the DataFrame
print(df.head())

# Calculate salary statistics using NumPy
average_salary = np.mean(df['Salary'])
std_dev_salary = np.std(df['Salary'])

# Print the calculated statistics
print(f'Average Salary: {average_salary}')
print(f'Standard Deviation of Salary: {std_dev_salary}')

# Create a bar chart of salary distribution
plt.hist(df['Salary'], bins=10, edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 讀取 CSV 檔案
data = pd.read_csv('adult.csv')

# 顯示數據
print(data.head())

# Calculate salary statistics using NumPy
average_salary = np.mean(data['Income']) 
std_dev_salary = np.std(data['Income'])  

# 選擇薪資欄位
salary_data = data['Income']

# 使用 NumPy 進行薪資統計
mean_salary = np.mean(salary_data)
std_salary = np.std(salary_data)
median_salary = np.median(salary_data)

# 顯示薪資統計結果
print('薪資的平均值為：', mean_salary)
print('薪資的標準差為：', std_salary)
print('薪資的中位數為：', median_salary)

# Print the calculated statistics
print(f'Average Salary: {average_salary}')
print(f'Standard Deviation of Salary: {std_dev_salary}')

# Create a bar chart of salary distribution
plt.hist(data['Income'], bins=10, edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.show()
```
